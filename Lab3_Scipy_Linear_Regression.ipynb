# Linregress does not work for multiple linear regression

```python
x_tv = x['TV']
x_radio = x['Radio']
x_news = x['Newspaper']

lr_tv = linregress(x_tv, y)
lr_radio = linregress(x_radio, y)
lr_news = linregress(x_news, y)

print("R2_Score for TV: ", lr_tv.rvalue)
print("R2_Score for Radio: ", lr_radio.rvalue)
print("R2_Score for Newspaper: ", lr_news.rvalue)
```

```python
X = dataset[['TV', 'Radio', 'Newspaper']]
X = sm.add_constant(X)
y = dataset['Sales']
model = sm.OLS(y, X).fit()
print("R2_score for 'Tv', 'Radio', 'Newspaper': ",model.rsquared)
print("\n",model.summary())
```
